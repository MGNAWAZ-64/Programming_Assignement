```python
#Write a Python program to find the area of a triangle?
#finding an area of triangle
#we will create a function 

def calculate_the_area_of_a_triangle(base,height):
    area = 0.5*base*height
    return area

base = float(input("Enter the value of base:"))
height = float(input("Enter the value of height:"))

area = calculate_the_area_of_a_triangle(base,height)
print(f"The area of the triangle is :{area}")

    
        
        
```

```python
#Write a Python program to swap two variables?
#Creating a function which accepts two variable
def swap_variable(a,b):
    try:
        temp = a
        a = b
        b = temp
        return a,b

    except Exception as e:
        print(e)
        
a = 5
b=6
print(f"Before swapping : a={a},b={b}.")
a,b = swap_variable(a,b)
print(f"After Swapping a = {a} and b={b}.")        
```

```python
#Write a Python program to generate a random number?
def generate_number(start,end):
    import random 
    return random.randint(start,end)

#Example usgae
start =int(input("Enter the number you want to start it from:"))
end =int(input("Enter the number you want to end:"))
random_number = generate_number(start,end)
print(f"The random variable generated by the compiler is : {random_number}")


    
```
